```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Excel file
file_path = 'Applei.xlsx'
xl = pd.ExcelFile(file_path)

# Load each sheet into a separate data frame
df_apple = xl.parse('Apple')
df_facebook = xl.parse('Facebook')
df_google = xl.parse('Google')



#df_apple 
#df_facebook 
#df_google
#
```

```python
df_apple.head(3)
```

```python
# DaTE TO Year

df_apple['Year_from_date'] = pd.to_datetime(df_apple['Date']).dt.year
df_facebook['Year_from_date'] = pd.to_datetime(df_facebook['Date']).dt.year
df_google['Year_from_date'] = pd.to_datetime(df_google['Date']).dt.year

```

```python
# Plot the line chart



sns.lineplot(data=df_apple, x='Year_from_date', y='Volume' ,label='Apple')
sns.lineplot(data=df_facebook, x='Year_from_date', y='Volume' , label='Facebook')
sns.lineplot(data=df_google, x='Year_from_date', y='Volume' , label='Google')




# Set the title and axes labels
plt.title('Volume')
plt.xlabel('Year')
plt.ylabel('Price')

plt.xlim(2016, 2020)

plt.ylim(0, 60000000)

# Show the legend
plt.legend()

# Display the chart
plt.show()
```

```python
import mplfinance as mpf

# Create a DataFrame with the relevant columns for the candlestick chart
df_candlestick = df_apple[['Date', 'Open', 'High', 'Low', 'Close']]

# Convert the 'Date' column to a datetime object and set it as the index
df_candlestick['Date'] = pd.to_datetime(df_candlestick['Date'])
df_candlestick.set_index('Date', inplace=True)

#start_date = datetime.datetime.strptime('22-08-2016', '%d-%m-%Y')
#end_date = datetime.datetime.strptime('27-08-2016', '%d-%m-%Y')


# Create the candlestick chart
mpf.plot(df_candlestick, type='candle', mav=(50, 200),  xlim=(start_date, end_date))
```

```python
import matplotlib.pyplot as plt

# Data
dividend = [df_apple['dividends'].sum(), df_google['dividends'].sum(), df_facebook['dividends'].sum()]
operating_expenses = [df_apple['operating expenses (in Millions)'].sum(), df_google['operating expenses (in Millions)'].sum(), df_facebook['operating expenses (in Millions)'].sum()]

# Labels
labels = ['Apple', 'Google', 'Facebook']

# Colors
colors = ['#ff9999','#66b3ff','#99ff99']

# Plot
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].pie(dividend, colors=colors, labels=labels, autopct='%1.1f%%', startangle=90)
ax[0].set_title('dividends')
ax[1].pie(operating_expenses, colors=colors, labels=labels, autopct='%1.1fM', startangle=90)
ax[1].set_title('operating expenses (in Millions)')

plt.show()
```
