```python
import pandas as pd
import numpy as np
```

```python
apple=pd.read_csv('./AAPL.csv')
facebook=pd.read_csv('./FB.csv')
google=pd.read_csv('./GOOGL.csv')

```

```python
apple["dividends"]=np.random.uniform(low=0.0, high=2.5, size=len(apple))
facebook["dividends"]=np.random.uniform(low=0.0, high=2.5, size=len(facebook))
google["dividends"]=np.random.uniform(low=0.0, high=2.5, size=len(google))
```

```python
apple["operating expenses (in Millions)"]=np.random.uniform(low=0.0, high=1.5, size=len(apple))
facebook["operating expenses (in Millions)"]=np.random.uniform(low=0.0, high=1.5, size=len(facebook))
google["operating expenses (in Millions)"]=np.random.uniform(low=0.0, high=1.5, size=len(google))
```

```python
dfs = [apple, facebook, google]
```

```python
for df in dfs:
    df['MA50'] = df.Close.rolling(50).mean()
    df['MA200'] = df.Close.rolling(200).mean()
```

```python
for df in dfs:
    df['Previous day close price'] = df.Close.shift(1)
```

```python
apple.head()
```

```python
for df in dfs:
    df['Change in price'] = df['Close'] - df['Previous day close price']
```

```python
apple.head()
```

```python
for df in dfs:
    df['Percent change in price'] = df.Close.pct_change()
```

```python
apple.head()
```

```python
for df in dfs:
    df['Previous day volume'] = df.Volume.shift(1)
```

```python
apple.head()
```

```python
for df in dfs:
    df['Change in volume'] = df['Volume'] - df['Previous day volume']
```

```python
apple.head()
```

```python
for df in dfs:
    df['Percent change in volume'] = df.Volume.pct_change()
```
